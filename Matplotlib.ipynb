Gráficos

```python
import pandas as pd
from seaborn import load_dataset
import matplotlib.pyplot as plt


# Cargar el dataset de iris
iris = load_dataset("iris")

# Renombrar columnas para reemplazar '.' por '_'
iris.columns = iris.columns.str.replace('.', '_', regex=False)

print(iris.head())
```

```python
# Contar muestras por especie
species_counts = iris['species'].value_counts()
print(species_counts)
```

 plt.subplots() es una función de matplotlib que crea una figura y un conjunto de ejes.Devuelve dos objetos:

* fig: Representa la figura completa, el contenedor global del gráfico (puede contener uno o más ejes).

* ax: Representa un eje dentro de la figura. Es donde realmente se trazan los gráficos (barras, líneas, puntos, etc.).

```python

# Configuración de la figura y el eje
fig, ax = plt.subplots(figsize=(8, 5))

# Crear diagrama de barras
ax.bar(species_counts.index, species_counts.values, color=['#1f77b4', '#ff7f0e', '#2ca02c'])

# Etiquetas y título
ax.set_title('Cantidad de muestras por especie', fontsize=14)
ax.set_xlabel('Especies', fontsize=12)
ax.set_ylabel('Cantidad', fontsize=12)

# Mostrar el gráfico
plt.show()
```

## Histograma

```python
fig, ax = plt.subplots(figsize=(8, 5))

# Histograma
ax.hist(iris['sepal_length'], bins=10, color='#2ca02c', edgecolor='black', alpha=0.7)

# Etiquetas y título
ax.set_title('Distribución de la longitud del sépalo', fontsize=14)
ax.set_xlabel('Longitud del sépalo (cm)', fontsize=12)
ax.set_ylabel('Frecuencia', fontsize=12)

# Mostrar el gráfico
plt.show()
```
